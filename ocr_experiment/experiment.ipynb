```python
from PyPDF2 import PdfFileWriter, PdfFileReader
import os
import sys
import ocrmypdf
from glob import glob
from pdf2image import convert_from_path

def split_pdf(input_file, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    
    input_pdf = PdfFileReader(input_file)
    for i, p in enumerate(input_pdf.pages):
        output = PdfFileWriter()
        output.addPage(p)
        with open(os.path.join(output_dir, f"page-{i:002d}.pdf"), "wb") as output_stream:
            output.write(output_stream)
            
            
def pdf_pages_to_images(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    files = glob(os.path.join(input_folder, "*.pdf"))
    for f in files:
        b = os.path.basename(f)
        images = convert_from_path(f)
        images[0].save(os.path.join(output_folder, b.replace(".pdf", ".jpg")))
            
def ocr_pages(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    files = glob(os.path.join(input_dir, "*.pdf"))
    for f in files:
        b = os.path.basename(f)
        ocrmypdf.ocr(f, os.path.join(output_dir, b), language="ara",
                     sidecar=os.path.join(output_dir, b.replace(".pdf", ".txt")))
        
    
```

```python
input_files = glob("data/inputs/*.pdf")
for f in input_files:
    b = os.path.basename(f)
    split_pdf(f, "data/inputs_pages/"+b.replace('.pdf', ''))
```

```python
from google.cloud import vision
from google.auth import load_credentials_from_file
from pdf2image import convert_from_path


credentials, project_id = load_credentials_from_file("service-account-key.json")
_vision_client = vision.ImageAnnotatorClient(credentials=credentials)

def google_ocr(pdf_file):
    images = convert_from_path(pdf_file)
    images[0].save("tmp.jpg")
    with open("tmp.jpg", 'rb') as image_file:
        content = image_file.read()

    image = vision.Image(content=content)
    response = _vision_client.document_text_detection(image=image)
    
    return response

def ocr_pages_2(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    files = glob(os.path.join(input_dir, "*.pdf"))
    for f in files:
        b = os.path.basename(f)
        output = google_ocr(f)
        with open(os.path.join(output_dir, b.replace(".pdf", ".txt")), "w") as ff:
            ff.write(output.text_annotations[0].description)
```
